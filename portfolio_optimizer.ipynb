```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
desired_stocks = input("Enter the stocks you would like to buy in the format 'Ticker Ticker Ticker...': ")
tickers_list = desired_stocks.split()
NUMBER_OF_TICKERS = len(tickers_list)
```

```python
def download_data(tickers, start, end):
  """Download desired data from yfinance in a specific period"""
  return yf.download(tickers, start=start, end=end, interval="1d")
```

```python
# Download train data for 5 stocks (AAPL, MSFT, TSLA, NVDA, AMZN).
train_data = download_data(desired_stocks, "2018-01-01", "2023-01-01")
train_data
```

```python
# Download test data for 5 stocks (AAPL, MSFT, TSLA, NVDA, AMZN).
today = datetime.datetime.today().strftime('%Y-%m-%d')
test_data = download_data(desired_stocks, "2023-01-01", today)
test_data
```

```python
# Download SPY for benchmarking
spy_data = download_data("SPY", "2023-01-01", today)
spy_data
```

```python
# Keep only the closing adjusted price
closing_train_data = train_data['Close']

# Get the change percentages
closing_train_data_pct_changes = closing_train_data.pct_change().dropna()
```

# 1st Approach: Monte Carlo sampling

```python
def calculate_annualized_metrics(returns, weights):
  """Find return, volatility and sharpe ratio of portfolio"""
  # Return rate
  annual_return_rate = returns.mean() * 252
  portfolio_annual_return_rate = np.dot(weights, annual_return_rate)

  # Volatility
  cov_matrix = returns.cov().to_numpy()
  daily_volatility = np.sqrt(weights.T @ cov_matrix @ weights)
  portfolio_annual_volatility = daily_volatility * np.sqrt(252)

  # Sharpe ratio
  sharpe_ratio = portfolio_annual_return_rate / portfolio_annual_volatility

  return (portfolio_annual_return_rate, portfolio_annual_volatility, sharpe_ratio)
```

```python
def monte_carlo_portfolios(number):
  """Create a number of random portfolios"""
  portfolios = []

  while len(portfolios) < number:
    # Create and normilize a random portfolio
    random_portfolio = np.random.random(NUMBER_OF_TICKERS)
    sum = random_portfolio.sum()
    normalized_random_portfolio = random_portfolio / sum

    # Find the metrics of the portfolio
    metrics = calculate_annualized_metrics(closing_train_data_pct_changes, normalized_random_portfolio)

    # Store the results in the list
    portfolios.append({
      "weights": normalized_random_portfolio,
      "return": metrics[0],
      "volatility": metrics[1],
      "sharpe": metrics[2]
    })
  return portfolios
```

```python
# Create 20000 random portfolios
random_portfolios = monte_carlo_portfolios(20000)
```

```python
len(random_portfolios)
```

```python
# Convert the list of dictionaries to a pandas DataFrame
portfolios_df = pd.DataFrame(random_portfolios)

# Find the portfolio with the highest Sharpe Ratio
best_portfolio = portfolios_df.loc[portfolios_df['sharpe'].idxmax()]

# Print the Sharpe Ratio of the best portfolio
print(f"Best Portfolio Sharpe Ratio: {best_portfolio['sharpe']}, with portfolio split:")
for i in range(len(tickers_list)):
  print(f"{tickers_list[i]}: {round(best_portfolio['weights'][i]*100, 1)}%")

# Plot the portfolios and highlight the best one
plt.figure(figsize=(10, 6))
plt.scatter(portfolios_df['volatility'], portfolios_df['return'], c=portfolios_df['sharpe'], cmap='viridis', marker='o')
plt.colorbar(label='Sharpe Ratio')
plt.scatter(best_portfolio['volatility'], best_portfolio['return'], color='red', marker='*', s=200, label='Best Portfolio')
plt.title('Efficient Frontier')
plt.xlabel('Volatility (Standard Deviation)')
plt.ylabel('Expected Return')
plt.legend()
plt.grid(True)
plt.show()
```

# 2nd Approach: Optimization with SLSQP method, assuming no shorting and no leverage allowed


```python
import scipy
```

```python
# Define the negative sharpe ratio function to use it for minimization
def negative_sharpe_ratio(weights):
  """Given a weigths array it computes the negative sharpe ratio"""
  metrics = calculate_annualized_metrics(closing_train_data_pct_changes, weights)
  return -metrics[2]
```

```python
def optimize_portfolio():
  """Optimize the portfolio with respect to sharpe ratio"""
  res = scipy.optimize.minimize(negative_sharpe_ratio, np.ones(NUMBER_OF_TICKERS)/NUMBER_OF_TICKERS, bounds=[(0, 1)] * NUMBER_OF_TICKERS, constraints={"type": "eq", "fun": lambda weights: np.sum(weights) - 1})
  return res
```

```python
# Minimize the negative sharpe ratio function
optimized_portfolio = optimize_portfolio()
```

```python
print(f"The best sharpe ratio for these stocks is {-optimized_portfolio.fun} with the following split: ")
for i in range(NUMBER_OF_TICKERS):
  print(f"{tickers_list[i]}: {round(optimized_portfolio.x[i]*100, 1)}%")
```

# 3rd Approach: More constraints (Maximum Allocation, Minimum Allocation, Minimum Target Retrun)

```python
max_allocation = float(input("Enter the maximum allocation per asset you want in the format of a number in [0, 1]: "))
min_allocation = float(input("Enter the minimum allocation per asset you want in the format of a number in [0, 1]: "))
min_target_return = float(input("Enter the minimum target return you want in the format of a number in [0, 1]: "))
```

```python
def optimize_portfolio_with_constraints(min_target_return, annual_return_rate, max_allocation=1, min_allocation=0):
  """Optimize the portfolio with respect to sharpe ratio and the user constraints"""
  res = scipy.optimize.minimize(negative_sharpe_ratio, np.ones(NUMBER_OF_TICKERS)/NUMBER_OF_TICKERS, bounds=[(min_allocation, max_allocation)] * NUMBER_OF_TICKERS, constraints=[{"type": "eq", "fun": lambda weights: np.sum(weights) - 1},
   {"type":"ineq", "fun": lambda weights: np.dot(weights.T, annual_return_rate.to_numpy()) - min_target_return}])
  return res
```

```python
optimized_portfolio_with_constraints = optimize_portfolio_with_constraints(min_target_return, closing_train_data_pct_changes.mean() * 252, max_allocation, min_allocation)
```

```python
print("\n--- Optimization (With Constraints) Approach Results ---")
if optimized_portfolio_with_constraints.success:
  print(f"Best Sharpe Ratio with Constraints: {-optimized_portfolio_with_constraints.fun:.4f}")
  print("Portfolio Split:")
  for i in range(NUMBER_OF_TICKERS):
    print(f"  {tickers_list[i]}: {round(optimized_portfolio_with_constraints.x[i]*100, 1)}%")
else:
  print("Optimization with constraints failed to find a solution.")
optimized_portfolio_with_constraints.message
```

# Backtesting

```python
def backtest_portfolio(test_data, portfolio):
  """Backtest the portfolio on the test_data and return the annualized return rate"""
  # Keep only the start and end date to find total return
  closing_test_data = test_data["Close"]
  starting_and_closing_value = pd.concat([closing_test_data.head(1), closing_test_data.tail(1)])

  # Find total return
  percentage_change = (starting_and_closing_value.iloc[1] - starting_and_closing_value.iloc[0]) / starting_and_closing_value.iloc[0]
  percentage_change = percentage_change.to_numpy()
  portfolio_percentage_change = portfolio.T @ percentage_change
  return round((portfolio_percentage_change / 3) * 100, 1)
```

```python
print(backtest_portfolio(test_data, optimized_portfolio_with_constraints.x))
```

# Benchmarking

```python
def compare_with_benchmarks(test_data, spy_data, portfolio):
  """Plot the portfolio performance in the same graph with SPY"""
  # Keep only the closing data
  closing_test_data = test_data["Close"]
  closing_spy_data = spy_data["Close"]

  # Make the prices percentage changes
  test_data_returns = closing_test_data.pct_change().dropna()
  spy_returns = closing_spy_data.pct_change().dropna()

  # Find the Compounded Growth
  portfolio_growth = (1+ test_data_returns @ portfolio).cumprod()
  spy_growth = (1 + spy_returns).cumprod()

  # Date alignment
  common_dates = portfolio_growth.index.intersection(spy_growth.index)
  portfolio_growth = portfolio_growth.loc[common_dates]
  spy_growth = spy_growth.loc[common_dates]

  # Plot the growth
  plt.figure(figsize=(10, 6))
  plt.plot(portfolio_growth.index, portfolio_growth, label='Optimized Portfolio')
  plt.plot(spy_growth.index, spy_growth, label='SPY Benchmark')
  plt.title('Portfolio Growth vs. SPY Benchmark')
  plt.xlabel('Date')
  plt.ylabel('Cumulative Growth')
  plt.legend()
  plt.grid(True)
  plt.show()

compare_with_benchmarks(test_data, spy_data, optimized_portfolio_with_constraints.x)
```

# Risk Metrics

```python
def calculate_risk_metrics(portfolio, returns, alpha=0.95):
  """Return the VaR, CVaR, Maximum Drawdown of the portfolio"""
  # Value ant Risk (VaR)
  var = -np.percentile(returns, (1-alpha)*100)

  # Conditional Value at Risk (CVaR)
  sum = 0
  count = 0
  sorted_returns = np.sort(returns)
  while count < len(sorted_returns) and sorted_returns[count] < var:
      sum += sorted_returns[count]
      count += 1
  cvar = -sum/count

  # Maximum Drawdown
  cumulative = (1 + returns).cumprod()
  rolling_peak = cumulative.cummax()
  drawdown = (cumulative - rolling_peak) / rolling_peak
  max_drawdown = min(drawdown)
  return(var, cvar, max_drawdown)
```

```python
# Evaluate the performance of the optimized portfolio on the training data
train_portfolio_returns = closing_train_data_pct_changes @ optimized_portfolio_with_constraints.x
train_annualized_return, train_annualized_volatility, train_sharpe_ratio = calculate_annualized_metrics(closing_train_data_pct_changes, optimized_portfolio_with_constraints.x)

print("\n--- Training Set Statistics (Optimized Portfolio) ---")
print(f"Annualized Return: {train_annualized_return:.4f}")
print(f"Annualized Volatility: {train_annualized_volatility:.4f}")
print(f"Sharpe Ratio: {train_sharpe_ratio:.4f}")

train_var, train_cvar, train_max_drawdown = calculate_risk_metrics(optimized_portfolio_with_constraints.x, train_portfolio_returns)
print(f"Value at Risk (VaR) (95%): {train_var:.4f}")
print(f"Conditional Value at Risk (CVaR) (95%): {train_cvar:.4f}")
print(f"Maximum Drawdown: {train_max_drawdown:.4f}")


# Evaluate the performance of the optimized portfolio on the testing data
closing_test_data = test_data["Close"]
test_data_returns = closing_test_data.pct_change().dropna()
test_portfolio_returns = test_data_returns @ optimized_portfolio_with_constraints.x

# Calculate annualized metrics for the test set (requires daily returns)
# Ensure the test_data_returns DataFrame has the same columns as the weights
test_data_returns_aligned = test_data_returns[tickers_list]
test_annualized_return, test_annualized_volatility, test_sharpe_ratio = calculate_annualized_metrics(test_data_returns_aligned, optimized_portfolio_with_constraints.x)

print("\n--- Test Set Statistics (Optimized Portfolio) ---")
print(f"Annualized Return: {test_annualized_return:.4f}")
print(f"Annualized Volatility: {test_annualized_volatility:.4f}")
print(f"Sharpe Ratio: {test_sharpe_ratio:.4f}")

test_var, test_cvar, test_max_drawdown = calculate_risk_metrics(optimized_portfolio_with_constraints.x, test_portfolio_returns)
print(f"Value at Risk (VaR) (95%): {test_var:.4f}")
print(f"Conditional Value at Risk (CVaR) (95%): {test_cvar:.4f}")
print(f"Maximum Drawdown: {test_max_drawdown:.4f}")
```
